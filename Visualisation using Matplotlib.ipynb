```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Line Plot

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
## Changing Line Aesthetics
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='-.',linewidth=2)
plt.show()
```

```python
#Adding two lines in the same Plot
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title('Two Line Plot')
plt.xlabel('Multiple Plots')
plt.ylabel('Red & Green')
plt.grid(True)
plt.show()
```

```python
## Adding Sub-Plot
plt.subplot(1,2,1)#plt.subplot(row,column,position)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)


plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)#plt.subplot(row,column,position)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)


plt.grid(True)
plt.show()
```

```python
plt.subplot(2,2,1)#plt.subplot(row,column,position)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.subplot(2,2,4)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)


plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,4,5,8,9,6]
plt.scatter(x,a)
plt.show()
```

```python
# Changing Marks Aesthetics
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,4,5,8,9,6]
plt.scatter(x,a,marker='*',c='g',s=100)# marker-design requires, c=color, s=size
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,4,5,8,9,6]
b=[7,2,5,9,3,2,1,4,7]
plt.scatter(x,a,marker='*',c='g',s=100)# marker-design requires, c=color, s=size
plt.scatter(x,b,marker='.',c='r',s=100)
plt.show()
```

```python
# Adding Sub-Plot
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,4,5,8,9,6]
b=[7,2,5,9,3,2,1,4,7]
plt.subplot(2,1,1)
plt.scatter(x,a,marker='*',c='g',s=100)# marker-design requires, c=color, s=size
plt.subplot(2,1,2)
plt.scatter(x,b,marker='.',c='r',s=100)
plt.show()
```

# Bar Plot

```python
student={'Bob':87,'Matt':56,'Sam':27}
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(student.keys(),student.values())
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title('Bar Plot')
plt.xlabel('Marks')
plt.ylabel('Names')
plt.grid(True)
plt.show()
```

# Histogram

```python
data=[1,3,3,3,9,9,5,4,4]
plt.hist(data,bins=20,color='r')#for Dataframe use (df['column'] in place of data)
plt.show()
```

# Box Plot and Violin Plot

```python
one=[1,2,3,4,7,3,4,2,5,7,8]
two=[3,6,7,4,2,1,7,9,4,0,6]
three=[4,5,6,3,2,2,3,3,6,7,8]
```

```python
data=[one,two,three]
```

```python
plt.boxplot(data)
plt.show()
```

```python
#Violin Plot
```

```python
plt.violinplot(data)
plt.show()
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

# Pie-Chart and Doughnut Chart

```python
fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# Changing Aesthetics
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','green','blue','pink'])
plt.show()
```

```python
# Doughnut Chart
```

```python
plt.pie(quantity,labels=fruit,radius=2,colors=['yellow','green','blue','pink'])
plt.pie([1],colors=['white'],radius=1)
plt.show()
```
